```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head(4)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income per cap')
plt.scatter(df.year,df[['per capita income (US$)']].values,color='green',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']].values,df[['per capita income (US$)']].values)
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income per cap')
plt.scatter(df.year,df[['per capita income (US$)']].values,color='green',marker='+')
plt.plot(df.year,reg.predict(df[['year']].values),color='red')
```
